```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class SimpleMLP(nn.Module):
    def __init__(self):
        super().__init__()
        # 定义最简单的MLP结构：输入层 -> 隐藏层 -> 输出层
        self.fc1 = nn.Linear(4, 2)  # 输入4个特征，隐藏层8个神经元
        #  在PyTorch的 nn.Linear中，权重矩阵的形状是 [out_features, in_features]
        self.fc2 = nn.Linear(2, 2)   # 输出2个类别
        # 可以添加一个简单的激活函数
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
 class SimpleMLP(nn.module):
        def __init__(self):
            super().__init__()
            self.model = nn.Sequential(
                    nn.Linear(4, 2),
                    nn.ReLU(),
                    nn.Linear(2,2))
            
            
        def forward(self, x):
            x=self.model(x)
            return x
```

```python
model = SimpleMLP()
print(model)
```

```python
total_params = 0
trainable_params = 0

for name, param in model.named_parameters():
    if param.requires_grad:
        print(f"参数名: {name}")
        print(f"参数形状: {param.shape}")
        print(f"参数数量: {param.numel()}")
        print(f"参数值:\n{param.data}")
        print("-" * 30)
        total_params += param.numel()
        trainable_params += param.numel()
```

```python
test_input = torch.randn(3, 4)
y = model(test_input)
print(y)
```
